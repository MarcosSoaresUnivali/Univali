```python
#--------------------------------------------------------------------------------------------------
#  Lista 2 – Questão 4 - Resposta em Frequencia para as funções transferencia da Questão 3.
#
#  a) H(z) =  3.     z - 1.2            b) H(z) =          z            c) H(z) =       z + 0.9
#               -------------------               ------------------               ---------------
#               (z - 0.5).(z - 0.9)               (z - 0.9)(z - 1.2)                z¨2 + z + 0.41 
#
#--------------------------------------------------------------------------------------------------
import matplotlib.pyplot as plt
from scipy import signal
```

```python
Fs = 8000
L = 8
```

```python
#--------------------------------------------------------------------------------------------------
#  a) H(z) =  3.     z - 1.2            =       3z - 3.6        =   0z¨2 + 3z - 3.6
#               -------------------       ------------------       -----------------
#               (z - 0.5).(z - 0.9)        z¨2 - 1.4 + 0.45         z¨2 - 1.4 + 0.45 
#
#--------------------------------------------------------------------------------------------------
fun = ' a)'
num = [0, 3, -3.6]
den = [1, -1.4, 0.45]
```

```python
#--------------------------------------------------------------------------------------------------
#  b) H(z) =          z            =       0z¨2 + z + 0       
#            -------------------        ------------------
#             (z - 0.9)(z - 1.2)         z¨2 - 2.1 + 1.08 
#
#--------------------------------------------------------------------------------------------------
fun = ' b)'
num = [0, 1, 0]
den = [1, -2.1, 1.08]
```

```python
#--------------------------------------------------------------------------------------------------
#  c) H(z) =    z + 0.9         =   0z¨2 + z + 0.9
#            ---------------        --------------
#             z¨2 + z + 0.41        z¨2 + z + 0.41
#
#--------------------------------------------------------------------------------------------------
fun = ' c)'
num = [0, 1, 0.9]
den = [1, 1, 0.41]
```

```python
delimi = ' '
my_num = delimi.join(str(num))
my_den = delimi.join(str(den))

w, H = signal.freqz(num, den, Fs)
Freq = w
plt.plot(Freq, abs(H))
plt.title('Resposta em frequencia para' + fun + ' :\n' + my_num.strip() + '\n--------------------------\n' + my_den.strip())
plt.show()
```
